```python
def run(date_ref: datetime, is_raw: bool = True) -> None:
    try:
        if date_ref.weekday() in (5, 6):
            logger.warning(f"📅 Ngày {date_ref.strftime('%d/%m/%Y')} là cuối tuần. Bỏ qua.")
            return None

        df = extractor.run(date_ref, is_raw=is_raw)
        if df is not None:
            return df
    except Exception as e:
        logger.error(f"🚨 Sự cố nghiêm trọng sập hệ thống toàn cục: {e}", exc_info=True)
```
